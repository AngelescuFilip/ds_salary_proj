```python
import pandas as pd 
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AngelescuFilip/ds_salary_proj/master/ai_jobs/ai_jobs_salaries.csv')
```

```python
df.head(20)
```

```python
df.columns
```

```python
def seniority(experience):
  if 'EN' in experience:
    return 'junior'
  elif 'MI' in experience:
    return 'intermediate'
  elif 'SE' in experience:
    return 'senior'
  elif 'EX' in experience:
    return 'director'
  else:
    return 'na'

def employment_type(employment):
  if 'PT' in employment:
    return 'part-time'
  elif 'FT' in employment:
    return 'full-time'
  elif 'CT' in employment:
    return 'contract'
  elif 'FL' in employment:
    return 'freelance'
  else:
    return 'na'
```

```python
df['exp_simp'] = df['experience_level'].apply(seniority)
```

```python
df.head(30)
```

```python
df['exp_simp'].value_counts()
```

```python
df['employment_simp'] = df['employment_type'].apply(employment_type)
```
